## Tesla stock price prediction

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from fbprophet import Prophet
%matplotlib inline
```

```python
# reading dataset
data = pd.read_csv("TSLA.csv")
```

```python
data.head()
```

```python
# visualizing relation between date and close feature
pl = data['Close'].plot(title='Tesla')
pl.set_xlabel('Date')
pl.set_ylabel('Close')
plt.show()
```

```python
# converting date column into datetime
data["Date"] = pd.to_datetime(data["Date"], infer_datetime_format=True)
data = data[["Date", "Close"]]
data = data.rename(columns={"Date" : "ds", "Close" : "y" })
```

```python
# creating an estimator: model using prophet instance
model = Prophet()
model.fit(data)
# making future dataset ie predict dataframe
predict = model.make_future_dataframe(periods=365)
# making predictions on predict being created and saving it to forecast
forecast = model.predict(predict)
```

```python
# visualizing prediction
model.plot(forecast,xlabel="Date", ylabel="Price")
plt.show()
```
