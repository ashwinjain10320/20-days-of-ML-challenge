## Google search analysis

```python
## importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
```

```python
## use of trends library.
## an API that can be used to access google data and analyse daily searches on google
```

```python
# creating an instance of trends
trends = TrendReq()
```

```python
# creating dataset on top 20 countries
trends.build_payload(kw_list=["Data Analytics"])
df = trends.interest_by_region()
df = df.sort_values(by="Data Analytics",ascending=False)
df = df.head(10)
df
```

```python
# visualising of frequency of searches from different regoins
df.reset_index().plot(x="geoName", y="Data Analytics", figsize=(20,15), kind="bar")
plt.show()
```

```python
## visualising yearly google searches over same keyword "data analytics"
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Data Analytics'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['Data Analytics'].plot()
plt.title('Total Google Searches for Data Analytics', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
